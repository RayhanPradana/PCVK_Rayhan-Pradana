```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pyplot as plt
```

```python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg"
        ]

for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi, "x", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

cv.line(image_3, (0, 0), (lebar - 1, tinggi - 1), (255, 255, 255))
cv.line(image_3, (0, tinggi - 1), (lebar - 1, 0), (255, 255, 255))

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

panjang_garis = int(2 * 100 / 2.54)
posisi_awal = (lebar - panjang_garis) // 2 - 20

for x in range(posisi_awal, posisi_awal + panjang_garis):
    image_3[int(tinggi / 2), x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Ukuran garis yang akan menjadi sisi-sisi kotak
panjang_garis = int(2 * 100 / 2.54)  # konversi dari cm ke pixel
posisi_awal_x = (lebar - panjang_garis) // 2 - 20  # posisi awal X untuk kotak
posisi_awal_y = (tinggi - panjang_garis) // 2  # posisi awal Y untuk kotak

# Menggambar kotak dan mengisi bagian dalamnya
for x in range(posisi_awal_x, posisi_awal_x + panjang_garis):
    for y in range(posisi_awal_y, posisi_awal_y + panjang_garis):
        image_3[y, x] = [255, 255, 255]  # Warna putih

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
